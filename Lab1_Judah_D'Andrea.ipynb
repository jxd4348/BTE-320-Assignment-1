## 2. Describe the Dataset

The **Healthy Lifestyle Cities Report 2021** compares 44 major cities worldwide on a range of factors linked to a healthy lifestyle. Each city receives an overall rank and score based on variables such as sunshine hours, life expectancy in the host country, air pollution levels, cost of water, average working hours, obesity prevalence, and access to outdoor spaces. The dataset also includes lifestyle cost indicators like monthly gym membership fees and weekly take‑out meal costs.

From our inspection using `str()`, `head()`, and `summary()`, we see that most columns are numeric (after cleaning), while the first column holds city names. Notable variables include:

- **Rank**: Overall position in the report’s healthiness ranking (1 = healthiest).
- **Sun**: Average annual sunshine hours in the city.
- **WaterCost**: Price of one bottle of water in USD.
- **Obesity**: Percentage of the population classified as obese.
- **LifeExp**: Average life expectancy (years) in the country where the city is located.
- **Pollution**: Air quality index (higher values indicate worse pollution).
- **WorkHours**: Average weekly working hours.
- **Happiness**: Score derived from the World Happiness Report (0–10 scale).
- **Outdoor**: Number of outdoor activities or facilities per capita.
- **Takeout**: Average weekly cost of ordering take‑out food.
- **GymCost**: Average monthly cost of a gym membership.

Overall, the dataset provides a broad, multi‑dimensional view of urban health and lifestyle factors across different regions, enabling exploratory analysis of how environment, cost of living, and cultural habits interplay in shaping well‑being.

# 8. Correlation Analysis

```r
# Correlation matrix of numeric variables
cor_matrix <- cor(numeric_df, use = "pairwise.complete.obs")
round(cor_matrix, 3)
```

### Interpretation of Correlations

- **Strong positive correlation:** `Sun` and `Happiness` show a high positive correlation, suggesting that cities with more sunshine hours tend to report higher happiness scores.  
- **Strong negative correlation:** `Pollution` and `LifeExp` display a strong negative correlation, indicating that higher pollution levels are associated with lower life expectancy.  
- **Other observations:** Some variables, such as `GymCost` and `Obesity`, exhibit weak correlations, implying that gym membership cost alone does not strongly relate to obesity rates across these cities.  
These patterns highlight how environmental and lifestyle factors interplay in influencing health outcomes.
