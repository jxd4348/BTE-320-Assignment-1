In class Assignment 1, 9/3

Final Code

```python
sales = input("Enter sales ($): ")
sales = float(sales)

cogs_pct = input("Enter cogs (%): ")
cogs_pct = float(cogs_pct)

rate = input("Enter tax rate (%): ")
rate = float(rate)

cogs = sales * (cogs_pct / 100)
grossProfit = sales - cogs
taxes = grossProfit * rate / 100
netIncome = grossProfit - taxes


print(f"cost of Goods sold: ${cogs}")
print(f"Gross Profit: ${grossProfit}")
print(f"Taxes: ${taxes}")
print(f"Net Income: ${netIncome}")
```

```python
import math

d = input("Enter Annual Demand (units/year): ")
d = float(d)

s = input("Enter Reorder cost ($/order): ")
s = float(s)

h = input("Enter holding cost($/unit/year): ")
h = float(h)

EOQ = (2 * d * s / h) ** (1/2)

print(f"annual demand: ${d}")
print(f"Reorder cost: ${s}")
print(f"Holding Cost: ${h}")
print(f"EOQ: ${EOQ}")
```

```python
x = float(input("Enter x: "))
y = float(input("Enter y: "))
op = input("Enter operator (+, -, *, /, //, **, %): ")

if op == "+":
    result = x + y
elif op == "-":
    result = x - y
elif op == "*":
    result = x * y
elif op == "/":
    result = x / y
elif op == "//":
    result = x // y
elif op == "**":
    result = x ** y
elif op == "%":
    result = x % y
else:
  result = None
  print("Invalid operator")

print(f"result = {result}")


```

Assignment 7 10/1

```python
#class version
import random

while True:

  play = input('Do you want to play a game? (yes/no): ').strip().lower()

  if play == 'no':
    break

  user_choice = input('Choose rock, paper, or scissor: ').strip().lower()
  pc_choice = random.choice(['rock', 'paper', 'scissor'])
  print(f'Computer chose: {pc_choice}')

  if user_choice == pc_choice:
    print('It\'s a tie!')
  elif (
    (user_choice == 'rock' and pc_choice == 'scissor') or
    (user_choice == 'paper' and pc_choice == 'rock') or
    (user_choice == 'scissor' and pc_choice == 'paper')
    ):
    print('User wins!')
  else:
    print('Computer wins!')
```

```python
def calcFee(h):
  fee = 5 + 2.5 * h
  if fee < 10.0:
    return 10.0
  elif fee > 20.0:
    return fee
  else:
    return fee


def main():
  hours = int(input('Enter number of parking hours: '))
  fee = calcFee(hours)
  print(fee)

if __name__ == '__main__':
  main()

```
