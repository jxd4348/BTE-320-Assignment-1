In Class Assignment 1, 9/3 (Income Statement)

Final Code

```python
sales = input("Enter sales ($): ")
sales = float(sales)

cogs_pct = input("Enter cogs (%): ")
cogs_pct = float(cogs_pct)

rate = input("Enter tax rate (%): ")
rate = float(rate)

cogs = sales * (cogs_pct / 100)
grossProfit = sales - cogs
taxes = grossProfit * rate / 100
netIncome = grossProfit - taxes


print(f"cost of Goods sold: ${cogs}")
print(f"Gross Profit: ${grossProfit}")
print(f"Taxes: ${taxes}")
print(f"Net Income: ${netIncome}")
```

Assignment 7 10/1 (Rock, Paper, Scissors)

```python
#class version
import random

while True:

  play = input('Do you want to play a game? (yes/no): ').strip().lower()

  if play == 'no':
    break

  user_choice = input('Choose rock, paper, or scissor: ').strip().lower()
  pc_choice = random.choice(['rock', 'paper', 'scissor'])
  print(f'Computer chose: {pc_choice}')

  if user_choice == pc_choice:
    print('It\'s a tie!')
  elif (
    (user_choice == 'rock' and pc_choice == 'scissor') or
    (user_choice == 'paper' and pc_choice == 'rock') or
    (user_choice == 'scissor' and pc_choice == 'paper')
    ):
    print('User wins!')
  else:
    print('Computer wins!')
```

Assignment 11 (Quality Control Manager (exceptions))

```python
def getValues(n: int) -> list:
  items = []
  for i in range(n):
    try:
      item = float(input('Enter a weight value: '))
    except Exception as e:
      item = float('nan')

    items.append(item)
  print(item)
  return items

def calcAverage(alist:list) -> float:
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float('nan')

def main():
  n = int(input('How many weights? '))
  if n < 0:
    raise ValueError('Number of items must be positive.')

  L = getValues(n)
  avg = calcAverage(L)
  print(f'Average weight: {avg} lbs.')

main()
```

Assignment 13 11/12 (Greeting Class)

```python
class Greeting:
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse = True):
    if reverse:
      print(f'{self.last_name}, {self.first_name}')
    else:
      print(self.first_name)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse = False)
```

Assignment 14 11/17 (Student Class)

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def getName(self):
    return self.__name

  def getCourses(self):
    return self.__courses

  def setName(self, newname):
    self.name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)
print(s.getName())
print(s.getCourses())
s.enroll('ABC123')
print(s.getCourses())
s.enroll('ABC123')
s.enroll('ABC157')
print(s.getCourses())
len(s)
```

Assignment 15 11/19 (Classes Point and Rectangle)

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    x_diff_sq = (self.x - other.x) ** 2
    y_diff_sq = (self.y - other.y) ** 2

    return (x_diff_sq + y_diff_sq) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0, 0)
    self.p2 = Point(1, 0)
    self.p3 = Point(1, 1)
    self.p4 = Point(0, 1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p4)

  def area(self):
    return self.width * self.height


sq = Rectangle()
sq.area()
```
